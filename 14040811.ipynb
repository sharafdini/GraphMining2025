```python
import networkx as nx

G = nx.Graph()   # null graph: no vertex no edges

V = list('abcdefg')
E = [('a','g'),('a','e'),('b','f'),('e','c'),('a','d'),('f','d')]

G.add_nodes_from(V)
G.add_edges_from(E)

nx.draw(G, with_labels=True)
```

```python
G.add_node('z')
nx.draw(G,with_labels=True,node_size=800,node_color='red',font_size='25',font_color='green')
```

```python
G.add_edge('z','a')
nx.draw(G,with_labels=True,node_size=800,node_color='red',font_size='25',font_color='green')
```
