```python
# nx.path_graph(4)

P4 = nx.path_graph(4)
nx.draw_circular(P4,with_labels=True,node_size=1000,node_color='lightblue',font_size='10',font_color='green',edgecolors='blue')
```

```python
# nx.cycle_graph(4)

C4 = nx.cycle_graph(4)
nx.draw_circular(C4,with_labels=True,node_size=1000,node_color='lightblue',font_size='10',font_color='green',edgecolors='blue')
```

```python
# nx.star_graph(5)

S5 = nx.star_graph(5)
nx.draw_circular(S5,with_labels=True,node_size=1000,node_color='lightblue',font_size='10',font_color='green',edgecolors='blue')
```

```python
# nx.complete_graph(5)
K5= nx.complete_graph(5)
nx.draw_circular(K5,with_labels=True,node_size=1000,node_color='lightblue',font_size='10',font_color='green',edgecolors='blue')
```
