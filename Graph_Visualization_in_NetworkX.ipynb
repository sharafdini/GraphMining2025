# Python Visualize Graphs Generated in NetworkX Using Matplotlib

This notebook demonstrates how to visualize graphs created using the **NetworkX** library and **Matplotlib**. We will cover various layout styles for representing the graph and show how to save the graph as an image.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Creating a Simple Graph

In this cell, we create a basic undirected graph and visualize it. The graph contains 4 nodes and 4 edges.

```python
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4)])

nx.draw(G, with_labels=True)
plt.show()
```

## Using Circular Layout

The **circular layout** arranges the nodes in a circle, which is a simple yet effective way to visualize smaller networks.

```python
nx.draw_circular(G, with_labels=True)
plt.title('Circular Layout')
plt.show()
```

## Using Planar Layout

The **planar layout** tries to arrange the graph without any edges crossing, which works well for certain types of graphs.

```python
nx.draw_planar(G, with_labels=True)
plt.title('Planar Layout')
plt.show()
```

## Using Random Layout

The **random layout** places the nodes at random positions within the graph, which can give a unique view for each visualization.

```python
nx.draw_random(G, with_labels=True)
plt.title('Random Layout')
plt.show()
```

## Using Spectral Layout

The **spectral layout** uses the eigenvectors of the graph's Laplacian matrix to position the nodes in a way that often emphasizes the structure of the graph.

```python
nx.draw_spectral(G, with_labels=True)
plt.title('Spectral Layout')
plt.show()
```

## Using Spring Layout

The **spring layout** (force-directed layout) positions the nodes as if they are connected by springs, leading to an aesthetically pleasing structure for most graphs.

```python
nx.draw_spring(G, with_labels=True)
plt.title('Spring Layout')
plt.show()
```

## Using Shell Layout

The **shell layout** arranges the nodes in concentric circles (shells), which can be useful for showing hierarchical or clustered structures.

```python
shell_layout = nx.shell_layout(G)
nx.draw_shell(G, nlist=[shell_layout], with_labels=True)
plt.title('Shell Layout')
plt.show()
```

## Saving the Visualization to a File

Finally, we demonstrate how to save the graph as an image file (PNG format).

```python
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True)
plt.title('Graph Saved as PNG')
plt.savefig('graph_image.png')
plt.close()
```
